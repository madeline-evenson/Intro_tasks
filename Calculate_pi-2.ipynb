```python
import random
import math
import matplotlib.pyplot as plt

total_counts = 10000

x_array = []
y_array = []

def pi_estimate():
    num_counts = 0
    num_inside_circle = 0
    for i in range(total_counts):
        x = random.random()
        y = random.random()
        num_counts += 1
        x_array.append(num_counts)
        distance = math.sqrt(x**2 + y**2)
        if distance < 1:
            num_inside_circle += 1
        pi = 4 * (num_inside_circle / total_counts)
        y_array.append(pi)
    return pi

print("The pi estimate for 10000 counts is:", pi_estimate())

plt.plot(x_array, y_array)
plt.xlabel("Number of Counts")
plt.ylabel("Pi Estimate")
plt.title("Pi Calculation")
```
